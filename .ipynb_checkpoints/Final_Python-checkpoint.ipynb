# Identifying sentiment
Here we define a function to pass the comments through that will extract a sentiment score.

```python
def sentiment_calc(df_final):
    try:
        return TextBlob(df_final).sentiment
    except:
        return None
```

The output of the sentiment function provides both a sentiment score and a subjectivity score, on a scale from -1 to 1.  Since we are only interested in sentiment, we use regular expressions to extract the sentiment score and store it as an integer.  We then store that integer as a list.

```python
df_final['sentiment'] = df_final['comments'].apply(sentiment_calc)

df_final['sentiment_String'] = df_final['sentiment'].astype(str)

df_final['sentiment_int'] = df_final.sentiment_String.str.extract('([-+]?\d*\.\d+|\d+)')

df_final['sentiment_int'] = df_final['sentiment_int'].astype(float)
```

```python
df_final
```

```python
sentiment_int = df_final["sentiment_int"].tolist()
```

In order to group the sentiment to be a categorical value we could relate comments to in our network graph, we create here a function to group the sentiment scores into positive, neutral, and negative classifications.  We then add this to the dataframe with an index of the columns.

```python
def get_comment_sentiment():
 sentiment = []
 for i in sentiment_int:
    if i > 0:
        sentiment.append('positive')
    elif i == 0:
        sentiment.append('neutral')
    else:
        sentiment.append('negative')
 return(sentiment)

comment_sentiment = get_comment_sentiment()
```

```python
df_final['sentiment'] = comment_sentiment
```

```python
df_final['index_col'] = df_final.index
```

```python
df_final
```

# Creating the graph
Here we create lists of the different nodes to have on hand.  We'll use them later to assign color to our graph.  Next, we assign the edges, connecting each comment to both a valence (positive, negative, or neutral) and to a source.

```python
node_categories = {
    "sentiment":df_final["sentiment"].tolist(),
    "source": df_final["source"].tolist(),
    "index_col": df_final["index_col"].tolist()
}
```

```python
for i in range(1):
    df_final["edges_1"] = df_final.apply(lambda x: (x["sentiment"], x["index_col"]), axis=1)
    df_final["edges_2"] = df_final.apply(lambda x: (x["sentiment"], x["source"]), axis=1)
    df_final["edges_3"] = df_final.apply(lambda x: (x["index_col"], x["source"]), axis=1)
```

```python
df_final
```

Here, we create a blank graph and add the edges and nodes, and then assign each node to its respective category (of either sentiment, source, or index, which is equivalent to comment observation in our dataset).

```python
G = nx.Graph()
G.add_edges_from(df_final["edges_1"].tolist()+df_final["edges_3"].tolist())
G.add_nodes_from(df_final["index_col"].tolist()+df_final["sentiment"].tolist() + df_final["source"].tolist())
for node in G:
    if node in df_final["sentiment"]:
        G.node[node]["category"] = "sentiment"
    elif node in df_final["source"]:
        G.node[node]["category"] = "source"
    else:
        G.node[node]["category"] = "index_col"
```

Here we build upon the graph by adding a spring layout and assigning the position of the nodes.  We then format the graph with customized coloring, font sizes, transparenty, line width, node shape, and node size.

```python
pos = nx.spring_layout(G,k=1.5,iterations=20)
pos_labels = {}
keys = pos.keys()
for key in keys:
    x, y = pos[key]
    pos_labels[key] = (x, y)
plt.figure(figsize=(30,30))
nx.draw_networkx_labels(G, pos_labels, font_color='black', font_size = 11, alpha=0.8)
nx.draw_networkx_edges(G, pos, edge_color="Gray", alpha=0.2)
nx.draw_networkx_nodes(G, pos, nodelist=node_categories["sentiment"], node_color= 'green', line_widths = 1.5, node_shape="h", node_size=900)
nx.draw_networkx_nodes(G, pos, nodelist=node_categories["index_col"], node_color="red", node_shape="o", node_size=300)
nx.draw_networkx_nodes(G, pos, nodelist=node_categories["source"], node_color="blue", linewidths = 1.5, node_shape="s", node_size=900)
plt.axis('off')
plt.show()
```
